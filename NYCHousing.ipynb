```python
# Standardize TYPE column by fixing typo 'Condop for sale'

df['TYPE'] = df['TYPE'].replace({'Condop for sale': 'Condo for sale'})
df['TYPE'].unique() # Check
```

```python
# Continue standardizing TYPE column
df['TYPE'] = df['TYPE'].str.replace('for sale','')
df['TYPE'] = df['TYPE'].str.replace('For sale','Undisclosed')
```

```python
df_numeric = df.loc[:, ['BEDS','BATH','PROPERTYSQFT','PRICE']]
df_numeric
```

```python
# Data split into X and Y

y = df_numeric['PRICE']
x = df_numeric.drop('PRICE', axis = 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y,
                                                    test_size = 0.2,
                                                    shuffle = True,
                                                    random_state = 42)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

# Model Building


## Linear Regression

```python
# Linear Regression

LR = LinearRegression()
LR = LR.fit(X_train, y_train)

# store the predictions
train_preds_LR = LR.predict(X_train)
test_preds_LR = LR.predict(X_test)
```

```python
# this is a regression problem, and there are three metrics to look at.
# 1) R2 value - for Linear Regression
print("This is train R2 (LR):", r2_score(y_train, train_preds_LR)) # train
print("This is test R2 (LR):", r2_score(y_test, test_preds_LR)) # test

# 2) MAE - for Linear Regression
trainMAE_LR = mean_absolute_error(y_train, train_preds_LR)
print("This is trainMAE (LR):", trainMAE_LR) # train
testMAE_LR = mean_absolute_error(y_test, test_preds_LR)
print("This is testMAE (LR):", testMAE_LR) # test

# 3) MSE
trainMSE_LR = mean_squared_error(y_train, train_preds_LR)
print("This is trainMSE (LR):", trainMSE_LR)
testMSE_LR = mean_squared_error(y_test, test_preds_LR)
print("This is testMSE (LR):", testMSE_LR)
```

```python
LR_train_r2_score = r2_score(y_train, train_preds_LR)
LR_test_r2_score = r2_score(y_test, test_preds_LR)
LR_trainMAE = mean_absolute_error(y_train, train_preds_LR)
LR_testMAE = mean_absolute_error(y_test, test_preds_LR)
LR_trainMSE = mean_squared_error(y_train, train_preds_LR)
LR_testMSE = mean_squared_error(y_test, test_preds_LR)
```

```python
LR_results = pd.DataFrame(['Linear_Regression', LR_train_r2_score, LR_test_r2_score, LR_trainMAE, LR_testMAE, LR_trainMSE, LR_testMSE]).transpose()
LR_results.columns = ['Method', 'Train R2','Test R2','Train MAE','Test MAE','Train MSE','Test MSE']
```

```python
LR_results
```

## Random Forest Regression

```python

DTR = DecisionTreeRegressor(min_samples_split=15)
# or start to 'tinker' :) this is called hyperparameter tuning
DTR = DTR.fit(X_train, y_train)

# store the predictions
train_preds_DTR = DTR.predict(X_train)
test_preds_DTR = DTR.predict(X_test)
```

```python
# 1) R2 value - for Decision Tree Regression
print("This is train R2 (DTR):", r2_score(y_train, train_preds_DTR)) # train
print("This is test R2 (DTR):", r2_score(y_test, test_preds_DTR)) # test

# 2) mae - for DT Regression
trainMAE_DTR = mean_absolute_error(y_train, train_preds_DTR)
print("This is trainMAE (DTR):", trainMAE_DTR) # train
testMAE_DTR = mean_absolute_error(y_test, test_preds_DTR)
print("This is testMAE (DTR):", testMAE_DTR) # test

# 3) mse
trainMSE_DTR = mean_squared_error(y_train, train_preds_DTR)
print("This is trainMSE (DTR):", trainMSE_DTR)
testMSE_DTR = mean_squared_error(y_test, test_preds_DTR)
print("This is testMSE (DTR):", testMSE_DTR)
```

```python
DTR_train_r2_score = r2_score(y_train, train_preds_DTR)
DTR_test_r2_score = r2_score(y_test, test_preds_DTR)
DTR_trainMAE = mean_absolute_error(y_train, train_preds_DTR)
DTR_testMAE = mean_absolute_error(y_test, test_preds_DTR)
DTR_trainMSE = mean_squared_error(y_train, train_preds_DTR)
DTR_testMSE = mean_squared_error(y_test, test_preds_DTR)
```

```python
DTR_results = pd.DataFrame(['Decision Tree', DTR_train_r2_score, DTR_test_r2_score, DTR_trainMAE, DTR_testMAE, DTR_trainMSE, DTR_testMSE]).transpose()
DTR_results.columns = ['Method', 'Train R2','Test R2','Train MAE','Test MAE','Train MSE','Test MSE']
```
